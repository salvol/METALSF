# Exercise 1 : Use xraydb python database to estimate absorption coefficient
## Using the code below compare absortion coefficient value of Fe at 20keV and 60keV

```python
import numpy as np
import matplotlib.pyplot as plt
import xraydb

# material and energy definition : energy is in eV
Material='Fe'
Energy=20000  # in eV

# compute mu value with xraydb in cm-1
muFe20=xraydb.material_mu(Material,Energy)
print('Material ', Material)
print('Energy in keV = ',Energy/1000)
print('mu (cm-1) = ', muFe20)

# material and energy definition : energy is in eV
Material='Fe'
Energy=60000  # in eV

# compute mu value with xraydb in cm-1
muFe60=xraydb.material_mu(Material,Energy)
print('Material ', Material)
print('Energy in keV = ',Energy/1000)
print('mu (cm-1) = ', muFe60)
```

# Exercise 3 :Use xraydb python database to estimate transmission for an energy range (between 10 keV and 100 keV) and select appropriate energy
## a) plot transmission as a function of energy for Fe and thickness of 0.1 cm 
## b) What minimum energy is required for a transmission ?

```python
plt.close('all')

# Material information
Material='Fe'
thickness=0.1 # in cm

# Energy range 10keV to 100keV
Emin=10000.0
Emax=100000.0
Estep = 1000.0
Energy_range=np.arange (Emin,Emax,Estep)

# compute transmission for a specific thickness using Beer Lambert Law for various energies
mu=xraydb.material_mu(Material,Energy_range)
trans=np.exp(-mu*thickness)

# plot transmission coefficient vs energy
plt.figure(1)

plt.plot(Energy_range/1000,trans,'b-')   
plt.xlabel('Energy in keV')
plt.ylabel('transmission')
title='Material = %s / thickness = %0.4f cm' %(Material,thickness)
plt.title(title)
```

#  Exercise 4 : Use xraydb python database to estimate transmission for a fixed energy and varying thickness 
## a) plot transmission as a function of thickness (ranging from 0.1 to 10 cm) for Fe and an energy of 60keV
## b) What is the expected transmission for 0.2 cm ?

```python
plt.close('all')

# Material information
Material='Fe'

# thickness range 0.1 to 10 cm
tmin=0.1
tmax=10.
tstep = 0.1
t_range=np.arange (tmin,tmax,tstep)

#Energy fied to 60 keV
Energy=60000.

# compute absorption coefficient for a fixed energy
mu=xraydb.material_mu(Material,Energy)

# compute transmission for a specific energy using Beer Lambert Law for various thicknesses
trans=np.exp(-mu*t_range)

# plot transmission coefficient vs energy
plt.figure(1)

plt.loglog(t_range,trans,'b-')   
plt.xlabel('thickness in cm')
plt.ylabel('transmission')
title='Material = %s / Energy = %f keV' %(Material,Energy/1000)
plt.title(title)
plt.ylim(0.01,1)
```

# Exercise 6 : Plot the contrast between Al and Cu as a function of energy : assume Al thickness is 1mm and Cu thickness is 50µm
## Contrast is defined as 

```python
plt.close('all')

# Material information
Material1='Al'
thickness1=0.1 # in cm
Material2='Cu'
thickness2=0.005 # in cm

# Energy range 10keV to 100keV
Emin=10000.0
Emax=100000.0
Estep = 100.0
Energy_range=np.arange (Emin,Emax,Estep)

# compute transmission for a specific thickness using Beer Lambert Law for various energies
mu1=xraydb.material_mu(Material1,Energy_range)
mu2=xraydb.material_mu(Material2,Energy_range)
trans=np.exp(-(mu1*thickness1+mu2*thickness2))

fig,ax1=plt.subplots()
ax2=ax1.twinx()

ax1.semilogy(Energy_range/1000,mu2-mu1,'b-')   
ax1.set_xlabel('Energy in keV')
ax1.set_ylabel('contrast between Al Cu')

ax2.plot(Energy_range/1000,trans,'g-')   
ax2.set_ylabel('transmission Al Cu')
```

## transmission values as a function of Energy and thickness for a specific material

```python
plt.close('all')

# Material information
Material='Fe'
thickness=0.15 # in cm

# Energy range 10keV to 100keV
Emin=10000.0
Emax=100000.0
Estep = 1000.0
Energy_range=np.arange (Emin,Emax,Estep)

#thickness range in cm
tmin = 0.01
tmax= 1
tstep=0.01
thickness=np.arange(tmin,tmax,tstep)


E, t = np.meshgrid(Energy_range, thickness)
valmu=np.ones((np.size(E,0),np.size(E,1)))
mu=xraydb.material_mu(Material,E[0,:])
valmu=valmu*mu[None,:]
trans=np.exp(-np.multiply(valmu,t))
levels = np.arange(0, 1, 0.05)
fig, ax = plt.subplots()
CS = ax.contour(E/1000,t*10,trans,levels)
ax.clabel(CS, inline=True, fontsize=10)
plt.xlabel('Energy in keV')
plt.ylabel('thickness in mm')
plt.title('transmission isovalue for Fe')
```
